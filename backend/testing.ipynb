```python
from llama_index import download_loader
```

```python
PDFReader = download_loader("PDFReader")
loader = PDFReader()
```

```python
from pathlib import Path
docs = loader.load_data(file=Path("C:/Users/artur/Downloads/MAGEE, Orthopedic Physical Assessment.pdf"))
```

```python
text = "\n\n".join([d.get_content() for d in docs])
```

```python
from llama_index import Document
```

```python
documents = [Document(text=text)]
```

```python
from llama_index.node_parser import SentenceSplitter
node_parser = SentenceSplitter(chunk_size=256, chunk_overlap=20)

base_nodes = node_parser.get_nodes_from_documents(documents=documents)
for i, node in enumerate(base_nodes):
    node.id_ = f"node-{i}"

for node in base_nodes:
    print(node.id_)
                                                  
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from llama_index.llms import OpenAI
from llama_index import OpenAIEmbedding
```

```python
gpt35 = OpenAI(model="gpt-3.5-turbo", api_key=os.getenv("OPENAI_API_KEY"))
embed_model = OpenAIEmbedding()
```

```python
from llama_index import ServiceContext
service_context = ServiceContext.from_defaults(llm=gpt35, embed_model=embed_model)
```

```python
from llama_index import VectorStoreIndex
base_index = VectorStoreIndex(base_nodes, service_context=service_context)
base_retriever = base_index.as_retriever(similarity_top_k=6)
retrievals = base_retriever.retrieve("What are some special tests for head and face injuries")
```
